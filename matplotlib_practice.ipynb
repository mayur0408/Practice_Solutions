```python
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df['species'].value_counts()
```

```python
mapping={'setosa':0,'virginica':1,'versicolor':2}
df_target=df['species'].replace({'setosa':0,'virginica':1,'versicolor':2}).copy()
```

```python
df_target.head()
```

```python
df1=df.drop(['species'],axis=1)
```

```python
final_dataset=pd.concat([df1,df_target], axis=1, ignore_index=True).copy()
final_dataset=final_dataset.rename(columns={0:'sepal_length',1:'sepal_width',2:'petal_length',3:'petal_width',4:'species'}).copy()
```

```python
final_dataset.head()
```

```python
final_dataset.corr()
```

```python
sns.heatmap(final_dataset.corr())
```

```python
sns.jointplot(x='sepal_length',y='species',data=final_dataset,kind='hex')
```

```python
sns.jointplot(x='sepal_length',y='species',data=final_dataset,kind='reg')
```

```python
sns.pairplot(final_dataset)
```

```python
sns.distplot(final_dataset['species'],kde=True)
```

```python
sns.countplot(final_dataset['species'])
```

```python
sns.barplot(x=final_dataset['species'],y=final_dataset['sepal_length'])
```
