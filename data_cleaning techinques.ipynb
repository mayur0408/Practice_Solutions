```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
pd.set_option('display.max_columns',None)
df=pd.read_csv('train.csv')
df.head()
```

```python
print(df.shape)
```

# Find Null Value Percentage

```python
features_with_na=[features for features in df.columns if df[features].isnull().sum()>1]
features_with_na

for features in features_with_na:
    print(f"{features} {np.round(df[features].isnull().mean(),4)} % missing values ")
```

# Find Relation between Missing values and target column

```python
for features in features_with_na:
    data=df.copy()
    data[features]=np.where(data[features].isnull(),1,0)
    data.groupby(data[features])['SalePrice'].mean().plot.bar()
    plt.title(features)
    plt.show()
```

 From this we can say that their exist direct relation between missing values and sale price !

# TO find Categorical Variable 

```python
categorical_feature=[features for features in df.columns if df[features].dtypes == 'O' ]
print(len(categorical_feature))
df[categorical_feature].head()
```

# Relation between Categorical Columns and target columns 

```python
for features in categorical_feature:
    df=data.copy()
    data.groupby(features)['SalePrice'].median().sort_values().plot.bar()
    plt.title(features)
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.show()
```

# Numerical Features 

```python
numerical_features=[features for features in df.columns if df[features].dtypes != 'O']
len(numerical_features)
```

```python
df[numerical_features].head()
```

```python
temporal_var=['YearBuilt','YearRemodAdd','GarageYrBlt','YrSold']
for features in temporal_var:
    if features != 'YrSold':
        data=df.copy()
        data[features]=data['YrSold']-data[features]
        plt.scatter(data[features],data['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_num=[features for features in numerical_features if len(df[features].unique())<25 and features not in temporal_var+categorical_feature+['ID']]
len(discrete_num)
```

```python
discrete_num
```

```python
for features in discrete_num:
    data=df.copy()
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.xlabel(features)
    plt.ylabel("SalePrice")
    plt.show()
```

```python
continuous_num=[features for features in numerical_features if features not in discrete_num+temporal_var+categorical_feature+['Id']]
```

```python
len(continuous_num)
```

```python
for features in continuous_num:
    data=df.copy()
    df[features].hist()
    plt.title(features)
    plt.xlabel(features)
    plt.show()
```
