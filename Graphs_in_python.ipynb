# Plotting graphs using matplotlib !

```python
%matplotlib inline 
from matplotlib import pyplot as plt #for displaying graph on notebook (magic function of jupyter nootbook !)
plt.title("Mayur's Data !") # for naming the graph !
plt.plot([1,2,3],[4,2,5]) #plotting single data on graph !
plt.ylabel("Y-axis") # for naming y axis !
plt.xlabel("X-axis") # for naming x axis !
plt.show()
```

```python
from matplotlib import style
a=[1,2,3]
b=[2,3,4]
a1=[4,5,6]
b1=[6,7,8]
plt.title("Graph for multiple data !")
plt.xlabel("This is X-axis")
plt.ylabel("This is Y-axis")
plt.plot(a,b,label="First",linewidth=3) #linewidth changes thickes of line for this data !
plt.plot(a1,b1,label="Second",linewidth=5)
plt.legend() # shows label of data on graph
style.use('ggplot') # Find more styles on matplotlib documentation !
plt.show()
```

```python
plt.title("Graph for different style plotting !")
plt.xlabel("This is X-axis")
plt.ylabel("This is Y-axis")
plt.scatter(a,b,label="First") #Find more plotting style on matplotlib documentation !
plt.hist(a1,b1,label="Second")
```
